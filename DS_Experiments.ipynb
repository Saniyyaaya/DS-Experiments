Using IQR to handle Outliers

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

**Encoding Categorical Variables:**Encoding categorical variables in data preprocessing involves transforming categorical data into numerical formats that machine learning algorithms can utilize.

```python
print(df.columns)
```

**One Hot Encoding:**One-hot encoding converts categorical values into a series of binary columns, where each column represents a unique category.

```python
df = pd.get_dummies(df, columns=['Funded Amount'])

df
```

**Label Encoding:**Label encoding converts categorical values into numerical values, assigning a unique integer to each category.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Grade']=le.fit_transform(df['Grade'])
df
```

**Ordinal Encoding:**Ordinal encoding replaces categorical values with numerical values based on their rank or order.

```python
from sklearn.preprocessing import OrdinalEncoder
size_mapping = {'tertiary':0, 'secondary':1, 'unknown':2}
df['Employment Duration']=df['Employment Duration'].map(size_mapping)
df
```

**Frequency Encoding:**Frequency encoding replaces each categorical value with its frequency (or count) in the dataset, helping to capture the distribution of categories.

```python
frequency_encoding = df['Loan Amount'].value_counts().to_dict()
df['Loan Amount'] =df['Loan Amount'].map(frequency_encoding)
df
```

**Target Encoding:**Target encoding replaces each categorical value with the mean of the target variable for that category, helping to capture the relationship between the categorical feature and the target variable.

```python
mean_encoded = df.groupby('Funded Amount Investor')['Interest Rate'].mean().to_dict()
df['Funded Amount Investor'] = df['Funded Amount Investor'].map(mean_encoded)
df
```

**Hashing Encoding:**Hashing encoding is a technique in data preprocessing where categorical data is transformed into a fixed-size numerical representation using a hash function. This is particularly useful for handling high-cardinality categorical features.

```python
df['Batch Enrolled'] = df['Batch Enrolled'].astype(str)
df
```

**Balancing the Dataset:**
Balancing the dataset is crucial when dealing with imbalanced classes. One common technique is to use Synthetic Minority Over-sampling Technique (SMOTE).

```python
from imblearn.over_sampling import RandomOverSampler

# Separate features and target variable
X = df.drop('Length', axis=1)  # Use the correct column name
y = df['Length']  # Use the correct column name

# Convert target variable to string
y = y.astype(str)

# Encode categorical variables
X = pd.get_dummies(X)

# Initialize and apply the RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Convert resampled data back to DataFrame
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['Length'] = y_resampled  # Use the correct column name

# Print the balanced dataset
print(df_resampled.head())
```
